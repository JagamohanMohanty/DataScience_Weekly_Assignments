### Task 1

```python
# Creating a function named operations that takes 2 inputs from the user
def operations(x,y):
    # Adding the two inputs
    print(f'Upon adding the two inputs we get the value to be {x+y}')
    # Substracting the two inputs
    print(f'Upon substracting the two inputs we get the value to be {x-y}')
    # Multiplying the two inputs
    print(f'Upon multiplying the two inputs we get the value to be {x*y}')
    # Dividing the two inputs
    print(f'Upon Dividing the two inputs we get the value to be {x/y}')
    # Modulus the two inputs
    print(f'Upon adding the two inputs we get the value to be {x%y}')
    # Exponation the two inputs
    print(f'Upon exponation the two inputs we get the value to be {x**y}')



# Printing results
operations(5,6)
```

### Task 2

```python
import random
# Setting the seed for reproducibility
random.seed(42)
# Creating a list of 10 random numbers
a = [random.randint(1,100) for _ in range(10)]
# The length of the list
print(f'The length of the list {len(a)}')
# The minimum and maximum element in the list
print(f'The minimum element in the list: {min(a)} and the maximum element in the list is {max(a)}')
# Removing one element from the list
print(f'The list before removing one element: {a}')
a.pop(6)
print(f'List after removing one element {a}')
# Adding one element to the list
a.insert(5,78)
print(f'List after adding one element {a}')
# Getting the list in ascending order
a.sort()
print(f'The list in ascending order {a}')
# Getting the list in descending order
a.sort(reverse=True)
print(f'The list in descending order {a}')
```

### Task 3

```python
# Creating the dictionary
student = {'name':'Jagamohan','age':22,'course':'Data Science','marks':89}
# Printing the key value pair
for k,v in student.items():
    print(f'key:{k} ---->>> value:{v}')

# Adding a new key
student['grade'] = 'A'
print(f'The new dict: {student}')

# Creating the set of unique courses
unique_courses = {'Python','Data Science','AI','Python'}
print(f'The unique courses set: {unique_courses}')

# Performing union, intersection and difference b/w 2 sets
set_A = {'Python','ML','AI'}
set_B = {'Data Science','Web Development','Python'}
print(f'Set A intersection Set B: {set_A.intersection(set_B)}')
print(f'Set A Union Set B: {set_A.union(set_B)}')
print(f'Set A difference Set B: {set_A.difference(set_B)}')
```
