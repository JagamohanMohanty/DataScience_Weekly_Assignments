### Exercise 1:  (Age Calculator)

```python
from datetime import datetime

try:
    birth_date_str = input("Enter your birth date (mm/dd/yyyy): ")
    birth_date = datetime.strptime(birth_date_str, "%m/%d/%Y")

    today = datetime.today()

    age = today.year - birth_date.year

    print(f"Your age is: {age} years")
    print("Birth date (European format):", birth_date.strftime("%d/%m/%Y"))

except ValueError:
    print("Error: Please enter date in mm/dd/yyyy format only.")

except Exception as e:
    print("Unexpected error occurred:", e)
```

### Exercise 2: (Prime Number Generator)

```python
# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

try:
    # Take range inputs
    start = int(input("Enter start of range: "))
    end = int(input("Enter end of range: "))

    # Validate positive integers
    if start <= 0 or end <= 0:
        raise ValueError("Inputs must be positive integers")

    # Store prime numbers
    primes = []

    # Find primes in range
    for num in range(start, end + 1):
        if is_prime(num):
            primes.append(num)

    # Display primes 
    print("Prime numbers:")
    for i in range(0, len(primes), 10):
        print(" ".join(map(str, primes[i:i+10])))

except ValueError as ve:
    print("Input Error:", ve)

except Exception as e:
    print("Unexpected error:", e)
```

### Example 3: Student Marks Processor

```python
with open("students.txt", "w") as file:
    file.write("REG001,78,65\n")
    file.write("REG002,55,60\n")
    file.write("REG003,88,90\n")
    file.write("REG004,45,50\n")
    file.write("REG005,67,72\n")
```

```python
students = []

try:
    with open("students.txt", "r") as f:
        for i, line in enumerate(f, 1):
            parts = line.strip().split(",")
            if len(parts) != 3:
                print(f"Skipping invalid line {i}: {line.strip()}")
                continue
            reg, exam_str, course_str = parts
            try:
                exam = float(exam_str)
                course = float(course_str)
            except ValueError:
                print(f"Skipping line {i} due to invalid numbers: {line.strip()}")
                continue
            overall = exam * 0.7 + course * 0.3
            grade = "A" if overall >= 70 else "B" if overall >= 60 else "C" if overall >= 50 else "F"
            students.append((reg, exam, course, overall, grade))

    if not students:
        print("No valid student data found.")
    else:
        students.sort(key=lambda x: x[3])

        with open("results.txt", "w") as f:
            for s in students:
                f.write(f"{s[0]} {s[3]:.2f} {s[4]}\n")

        grades_count = {}
        for s in students:
            grades_count[s[4]] = grades_count.get(s[4], 0) + 1

        print("Grade Statistics:")
        for grade in sorted(grades_count):
            print(f"{grade}: {grades_count[grade]}")

except FileNotFoundError:
    print("File error: students.txt not found.")
except Exception as e:
    print("Unexpected error:", e)
```

### Exampl 4: (Weather data fetcher & Analyzer)

```python
import requests
import csv

API_KEY = "31bd7505d23e87b9cdef4d0c16ce2cec"

def fetch_weather(city, api_key):
    try:
        url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
        response = requests.get(url)
        if response.status_code != 200:
            return None
        return response.json()
    except:
        return None

def analyze_weather(weather_data):
    temp = weather_data['main']['temp']
    wind = weather_data['wind']['speed']
    humidity = weather_data['main']['humidity']

    if temp <= 10:
        summary = "Cold (<=10 C)"
    elif 11 <= temp <= 24:
        summary = "Mild (11-24 C)"
    else:
        summary = "Hot (>=25 C)"

    if wind > 10:
        summary += " | High wind alert!"
    if humidity > 80:
        summary += " | Humid conditions!"

    return summary

def log_weather(city, filename):
    weather_data = fetch_weather(city, API_KEY)

    if weather_data is None:
        weather_data = {
            "main": {"temp": 33, "humidity": 78},
            "wind": {"speed": 9}
        }

    summary = analyze_weather(weather_data)

    with open(filename, "a", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow([
            city,
            weather_data['main']['temp'],
            weather_data['wind']['speed'],
            weather_data['main']['humidity'],
            summary
        ])

    print("Weather data logged successfully")
    print("Summary:", summary)


log_weather("Bhubneswar,IN", "weather_log.csv")
```
